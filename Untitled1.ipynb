# Análise de Produção de Petróleo no ano de 2021

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
dados = pd.read_csv('/content/anuario-2022-dados-abertos-produção-de-petroleo.csv', sep = ';')
dados.head()
```

```python
dados.shape
```

```python
dados.describe
```

```python
dados['PRODUÇÃO DE PETRÓLEO'] = dados['PRODUÇÃO DE PETRÓLEO'].apply(lambda x: str(x).replace(",","."))
```

```python
dados['PRODUÇÃO DE PETRÓLEO']
```

```python
dados['PRODUÇÃO DE PETRÓLEO'].astype('float64')
```

```python
dados = dados[['PAÍS','ANO','REGIÃO GEOGRÁFICA','PRODUÇÃO DE PETRÓLEO']]
dados.head(10)
```
